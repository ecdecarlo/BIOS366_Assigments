## Question 2

Carlin (1992) considers a Bayesian approach to meta-analysis, and includes the following examples of 22 trials of beta-blockers to prevent mortality after myocardial infarction. These data are given below.

In one possible random effects model we assume the true effect (on a log-odds scale) $d_i$ in a trial $i$
is drawn from some population distribution. Let $r^C_i$ denote number of events in the control group in trial $i$, and $r^T_i$ denote events under active treatment in trial $i$. Our model is:

$$\begin{aligned}
r^C_i &\sim \text{Binomial}\left(p^C_i, n^C_i\right) \\
r^T_i &\sim \text{Binomial}\left(p^T_i, n^T_i\right) \\
\text{logit}\left(p^C_i\right) &= \mu_i \\
\text{logit}\left(p^T_i\right) &= \mu_i + \delta \\
\mu_i &\sim \text{Normal}(m, s).
\end{aligned}$$

In this case, we want to make inferences about the population effect $m$, and the predictive distribution for the effect $\delta_{\text{new}}$ in a new trial. 

This particular model uses a random effect for the population mean, and a fixed effect for the treatment effect. There are 3 other models you could fit to represent all possible combinations of fixed or random effects for these two parameters.

Build all 4 models to estimate the treatment effect in PyMC and 

1. use convergence diagnostics to check for convergence in each model 
2. use posterior predictive checks to compare the fit of the models
3. use DIC to compare the models as approximations of the true generating model

Which model would you select and why?

```python
import pymc as pm
import numpy as np
```

```python
r_t_obs = [3, 7, 5, 102, 28, 4, 98, 60, 25, 138, 64, 45, 9, 57, 25, 33, 28, 8, 6, 32, 27, 22]
n_t_obs = [38, 114, 69, 1533, 355, 59, 945, 632, 278,1916, 873, 263, 291, 858, 154, 207, 251, 151, 174, 209, 391, 680]
r_c_obs = [3, 14, 11, 127, 27, 6, 152, 48, 37, 188, 52, 47, 16, 45, 31, 38, 12, 6, 3, 40, 43, 39]
n_c_obs = [39, 116, 93, 1520, 365, 52, 939, 471, 282, 1921, 583, 266, 293, 883, 147, 213, 122, 154, 134, 218, 364, 674]
N = len(n_c_obs)
```

## Part 1: Random effect for treatment, and fixed for population mean

```python
mu = pm.Normal('mu', .5, 16)
delta_T = pm.Normal('delta_T', 0, 4)
```

```python
def logit(p):
    return np.log(p) - np.log(1 - p)

def inv_logit(p):
    return np.exp(p) / (1 + np.exp(p))
```

```python
@pm.deterministic(plot=False)
def p_i(m=mu,d=delta_T):
    ''' Concatenate Poisson means '''
    pi_t = np.empty(len(r_t_obs))
    pi_t[::] = inv_logit(m+d)
    pi_c = np.empty(len(r_c_obs))
    pi_c[::] = inv_logit(m)
    return [pi_t,pi_c]
```

```python
[ri_t,ri_c] = pm.Binomial('ri_t', n=[n_t_obs,n_c_obs],p=p_i, 
                       value=[r_t_obs,r_c_obs], 
                       observed=True)
```

```python
M=pm.MCMC([mu, delta_T, p_i, ri_t])
```

```python
M.sample(iter=10000, burn=1000)
```

```python
from pymc.Matplot import plot
plot(M)
```

```python
mu.summary()
```

```python
delta_T.summary()
```

## Part 2: Fixed effect for treatment, and random for control

```python
mu = pm.Normal('mu', .5, 16)
delta_C = pm.Normal('delta_C', 0, 16)
```

```python
@pm.deterministic(plot=False)
def p_i(m=mu,d=delta_C):
    ''' Concatenate Poisson means '''
    pi_t = np.empty(len(r_t_obs))
    pi_t[::] = inv_logit(m)
    pi_c = np.empty(len(r_c_obs))
    pi_c[::] = inv_logit(m+d)
    return [pi_t,pi_c]
```

```python
[ri_t,ri_c] = pm.Binomial('ri_t', n=[n_t_obs,n_c_obs],p=p_i, 
                       value=[r_t_obs,r_c_obs], 
                       observed=True)
```

```python
M=pm.MCMC([mu, delta_C, p_i, ri_t])
M.sample(iter=10000, burn=1000)
```

```python
from pymc.Matplot import plot
plot(M)
```

## Part 3: Fixed effect for treatment, and fixed for control

```python
mu_c = pm.Normal('mu_c', .5, 16)
mu_t = pm.Normal('mu_t', .5, 16)

@pm.deterministic(plot=False)
def p_i(m_c=mu_c,m_t = mu_t):
    ''' Concatenate Poisson means '''
    pi_t = np.empty(len(r_t_obs))
    pi_t[::] = inv_logit(m_t)
    pi_c = np.empty(len(r_c_obs))
    pi_c[::] = inv_logit(m_c)
    return [pi_t,pi_c]
```

```python
[ri_t,ri_c] = pm.Binomial('ri_t', n=[n_t_obs,n_c_obs],p=p_i, 
                       value=[r_t_obs,r_c_obs], 
                       observed=True)
```

```python
M=pm.MCMC([mu_c, mu_t, p_i, ri_t])
M.sample(iter=10000, burn=1000)
```

```python
from pymc.Matplot import plot
plot(M)
```

## Part 4: random effect for treatment, and random for control

```python
Delta_c = pm.Normal('Delta_c', 0, 4)
Delta_t = pm.Normal('Delta_t', 0, 4)
mu = pm.Normal('mu',0.5,16)

@pm.deterministic(plot=False)
def p_i(d_t=Delta_t,d_c=Delta_c,m=mu):
    ''' Concatenate Poisson means '''
    pi_t = np.empty(len(r_t_obs))
    pi_t[::] = inv_logit(m+d_t)
    pi_c = np.empty(len(r_c_obs))
    pi_c[::] = inv_logit(m+d_c)
    return [pi_t,pi_c]
```

```python
[ri_t,ri_c] = pm.Binomial('ri_t', n=[n_t_obs,n_c_obs],p=p_i, 
                       value=[r_t_obs,r_c_obs], 
                       observed=True)
```

```python
M=pm.MCMC([Delta_t, Delta_c, mu, ri_t])
M.sample(iter=10000, burn=1000)
```

```python
from pymc.Matplot import plot
plot(M)
```

** Results Discussion **

The above results demonstrate the power of pymc and the effectiveness of representing the population effect and treatment and/or control effects in different ways. For instance, for the first three implementations we see good convergence, adequate coverage of the parameter space, and relativelylitle auto correlation in 10k samples. These three implementations cover the case where we treat the control as random, the treatment as random, or both as fixed. We see that the Bayesian methodology is able to parse out true and accurate representations of the population mean and one random effect, and also if both treament and control are assumed to have different means the Bayesian method similarly can deduce each. However, if we assume both control and treatment groups exhibit random effects, the calibration breaks down. For instance, we fail to get reasonable coverage of the parameter space and we continue to have high correlation through the sample. Also, the distributions are irregular and wide.

These results suggest that treating the control group and the treatment group as having random effects is not an accurate representation of the clinical trials.
